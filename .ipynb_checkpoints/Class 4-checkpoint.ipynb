```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, item):  # only applies to tail of list
        if not self.head:
            self.head = Node(item)
        else: 
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(item)
    
    def __str__(self):
        out = '['
        if self.head:
            out += '%s' % self.head.data
        current = self.head.next
        while current:
            out += ', %s ' % current.data
            current = current.next
        out += ']'
        return out
    
    #----------------------------------------------------------------------------------------
    
    def insert(self, index, item): # can insert data at any point of list
        if not self.head:
            self.head = Node(item)
        else:
            if index == 0:
                node = Node(item)
                temp = self.head
                self.head = node
                node.next = temp
            else:
                current = self.head
                prev = None
                counter = 0
                while current.next and counter != index:
                    prev = current
                    current = current.next
                    counter = counter + 1
                if counter == index:
                    node = Node(item)
                    prev.next = node
                    node.next = current
                else:
                    node = Node(item)
                    current.next = node
                    
    
    def remove(self, value):
        current =  self.head
        prev = None
        while current.next and current.data != value:
            prev = current
            current = current.next
        if current.data == value:
            if prev:
                temp = prev
                temp.next = current.next
            else:
                if not current.next:
                    self.head = current.next
        else:
            raise ValueError('The target is not present')
        #should be competency report
        
    
```
